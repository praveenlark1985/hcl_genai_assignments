```python
import matplotlib.pyplot as plt
```

```python
def draw_neural_network(layer_sizes):
    fig, ax = plt.subplots()

    n_layers = len(layer_sizes)
    v_spacing = 1
    h_spacing = 2

    # Draw neurons
    for i, layer_size in enumerate(layer_sizes):
        x = i * h_spacing
        y_top = (layer_size - 1) / 2

        for j in range(layer_size):
            y = y_top - j
            circle = plt.Circle((x, y), 0.2, fill=False)
            ax.add_artist(circle)

    # Draw connections
    for i in range(n_layers - 1):
        for j in range(layer_sizes[i]):
            for k in range(layer_sizes[i + 1]):
                x1 = i * h_spacing
                y1 = (layer_sizes[i] - 1) / 2 - j

                x2 = (i + 1) * h_spacing
                y2 = (layer_sizes[i + 1] - 1) / 2 - k

                ax.plot([x1, x2], [y1, y2])

    # Labels
    layer_names = ["Input Layer", "Hidden Layer 1", "Hidden Layer 2", "Output Layer"]
    for i, name in enumerate(layer_names):
        ax.text(i * h_spacing, -4, name, ha='center')

    # Feature names for input layer
    features = [
        "TransactionAmount",
        "TransactionTime",
        "MerchantCategory",
        "CustomerAge",
        "AccountBalance",
        "TransactionsToday"
    ]

    for i, feature in enumerate(features):
        ax.text(-1.5, 2.5 - i, feature, fontsize=9)

    # Output label
    ax.text((n_layers - 1) * h_spacing + 1, 0, "Fraud (0/1)", fontsize=10)

    ax.axis('off')
    plt.title("Neural Network Architecture for Fraud Detection")
    plt.show()
```

```python

# Define architecture: [Input, Hidden1, Hidden2, Output]
layer_sizes = [6, 8, 6, 1]

draw_neural_network(layer_sizes)
```
